```python
# Score atingindo no Codenation 99.34%
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

#import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
data_train = pd.read_csv("train.csv", sep=',', encoding='UTF8')
data_test = pd.read_csv("test.csv", sep=',', encoding='UTF8')
print( "successo!")
```

```python
data_train.head()
```

```python
data_train.isnull().sum() #verificação se há valores nulos
```

```python
# Verifica a correlação
corr = data_train.corr()['IN_TREINEIRO'].dropna() #.sort_values(ascending=False)
print(corr.head(15))
```

```python
features = [
    'TP_ST_CONCLUSAO',
    'TP_ANO_CONCLUIU',
    'NU_IDADE']

#features_corr = [
#    'IN_TREINEIRO',
#    'TP_ST_CONCLUSAO',
#    'TP_ANO_CONCLUIU',
#    'NU_IDADE']
```

```python
x = data_train[features]
y = data_train['IN_TREINEIRO']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
x_train
```

```python
svm=SVC(random_state=1)
svm.fit(x_train,y_train)
print("train accuracy:",svm.score(x_train,y_train))
print("test accuracy:",svm.score(x_test,y_test))
```

```python
data_to_predict = data_test[features]
y_pred_test = svm.predict(data_to_predict)
answer_csv = pd.DataFrame({'NU_INSCRICAO': data_test.NU_INSCRICAO, 'IN_TREINEIRO': y_pred_test})
answer_csv = answer_csv.round(2)
answer_csv.to_csv('answer.csv', index=False, header=True)

answer_csv.head()
```
